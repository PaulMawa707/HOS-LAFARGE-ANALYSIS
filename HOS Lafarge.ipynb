```python
import pandas as pd

# Load the Excel file
excel_file_path = 'HOS Violations Report - Bikash.xlsx'
excel_data = pd.ExcelFile(excel_file_path)

# Initialize an empty list to hold the data frames
data_frames = []

# Iterate through each sheet in the Excel file
for i, sheet_name in enumerate(excel_data.sheet_names):
    print(f"Processing sheet {i+1}/{len(excel_data.sheet_names)}: {sheet_name}")
    
    # Read the data starting from row 8
    sheet_data = pd.read_excel(excel_file_path, sheet_name=sheet_name, header=7)
    
    # Add a new column with the driver's name
    sheet_data['Driver'] = sheet_name
    
    # Append the DataFrame to the list
    data_frames.append(sheet_data)

# Concatenate all DataFrames once at the end
merged_data = pd.concat(data_frames, ignore_index=True)

# Save the merged data to a new Excel file
output_file = 'merged_drivers_data.xlsx'
merged_data.to_excel(output_file, index=False)

print(f"Data has been merged and saved to '{output_file}'.")
```

```python
import pandas as pd
import os

# Folder path containing Excel files
folder_path = 'Lafarge August HOS data'

# Initialize an empty list to store DataFrames
all_data = []

# Iterate through each Excel file in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.xlsx') or file_name.endswith('.xls'):
        file_path = os.path.join(folder_path, file_name)
        
        # Read all sheets in the Excel file
        xls = pd.ExcelFile(file_path)
        
        for sheet_name in xls.sheet_names:
            # Read data starting from the header row (row 8 in 0-indexed pandas is row 7)
            df = pd.read_excel(file_path, sheet_name=sheet_name, header=7)
            
            # Remove the last 5 rows (tail)
            df = df.iloc[:-5, :]
            
            # Add the driver column with the sheet name
            df['driver'] = sheet_name
            
            # Read the specific row (Row 6, 0-indexed is 5) to extract the site name
            site_info = pd.read_excel(file_path, sheet_name=sheet_name, header=None, nrows=6)
            # Extract the value from merged cell E to L in row 6
            site_full_text = site_info.iloc[5, 4]  # Column E in 0-index is column 4
            
            # Extract the site name after 'Controltech - '
            site_name = site_full_text.split('Controltech - ')[-1].strip() if 'Controltech - ' in site_full_text else site_full_text
            
            # Add the site name to the site column
            df['site'] = site_name
            
            # Append the cleaned DataFrame to the list
            all_data.append(df)

# Merge all DataFrames into one
merged_df = pd.concat(all_data, ignore_index=True)

# Export the merged DataFrame to a new Excel file
output_file = 'Merged_Lafarge_August_HOS_data.xlsx'
merged_df.to_excel(output_file, index=False)

print(f"Data merged and saved to {output_file}")
```

```python
df_cleaned = merged_df[merged_df['Start Date & Time'].notna() & (merged_df['Start Date & Time'].str.strip() != '')]
df_cleaned.head()
```

```python
df_cleaned.to_excel('cleaned_data.xlsx', index=False)
```

```python
# Convert 'Start Date & Time' column to datetime format with dayfirst=True
df_cleaned['Start Date & Time'] = pd.to_datetime(df_cleaned['Start Date & Time'], dayfirst=True, errors='coerce')

# Extract the day of the month and create a new 'day' column
df_cleaned['day'] = df_cleaned['Start Date & Time'].dt.day
df_cleaned.head()
```

```python
selected_columns =['Start Date & Time', 'day', 'End Date & Time', 'Type', 'Driving Time in Violation (hh:mm:ss)', 'Time In Violation (hh:mm:ss)', 'Violation Limit', 'driver', 'site']
df_cleaned = df_cleaned[selected_columns]
df_cleaned.head()
```

```python
# Convert 'Time In Violation (hh:mm:ss)' to timedelta, then to hours, and round to one decimal place
df_cleaned['Time in Violation (hrs)'] = (
    pd.to_timedelta(df_cleaned['Time In Violation (hh:mm:ss)'])
    .dt.total_seconds() / 3600
).round(1)

# Display the DataFrame with the new column
df_cleaned.head()
```

```python
import pandas as pd
from openpyxl import Workbook
from openpyxl.worksheet.table import Table, TableStyleInfo
from openpyxl.utils.dataframe import dataframe_to_rows

# Create a new Excel workbook and select the active worksheet
wb = Workbook()
ws = wb.active

# Append DataFrame rows to the worksheet
for r_idx, row in enumerate(dataframe_to_rows(df_cleaned, index=False, header=True), 1):
    for c_idx, value in enumerate(row, 1):
        ws.cell(row=r_idx, column=c_idx, value=value)

# Create a table and add it to the worksheet
table = Table(displayName="Table1", ref=f"A1:{chr(65 + len(df_cleaned.columns) - 1)}{len(df_cleaned) + 1}")

# Add a TableStyleInfo to the table
style = TableStyleInfo(
    name="TableStyleMedium9",
    showFirstColumn=False,
    showLastColumn=False,
    showRowStripes=True,
    showColumnStripes=True
)
table.tableStyleInfo = style

# Add the table to the worksheet
ws.add_table(table)

# Save the workbook
wb.save("df_cleaned_table.xlsx")
```
